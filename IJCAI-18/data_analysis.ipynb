### 读取训练数据

```python
# 选择特征
select_feature = ['item_price_level','item_sales_level','item_collected_level',\
                  'item_pv_level','user_gender_id','user_age_level','user_star_level',\
                  'context_page_id','shop_review_num_level','shop_review_positive_rate',\
                  'shop_star_level','shop_score_service','shop_score_delivery','shop_score_description']
print('number of features: %d'%len(select_feature))
```

```python
# 读取训练数据
import pandas as pd
train_data = pd.read_csv('data/round1_ijcai_18_train_20180301.txt', sep=' ')
train_data.head()
# 去重
train_data = train_data.drop_duplicates(select_feature)
```

## 线性回归分析

### 选择特征

```python
#model = pd.ols(y=train_data['item_id', 'item_category_list'], x=train_data['is_trade']) 

# 选择特征
train_data_x = train_data[select_feature]
train_data_y = train_data[['is_trade']]
train_data_x.head()
```

### 模型训练

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(train_data_x, train_data_y)
```

### 读取测试数据

```python
test_data = pd.read_csv('data/demo_test.txt', sep=' ')
# 选择特征
test_data_x = test_data[select_feature]
test_data_x.head()
```

### 测试

```python
result = linreg.predict(test_data_x)
result
```

### 结果融合

```python
resultdf =  pd.DataFrame(result, columns = ['predicted_score'])
tbresult = pd.concat([test_data[['instance_id']],resultdf], axis=1)
# tbresult = pd.DataFrame({test_data['instance_id'].data, result})
tbresult.head()
```

### 存储

```python
tbresult.to_csv('data/demo_result.txt', sep = ' ',index = False)
```

```python
# 计算训练集log_loss
train_predict = linreg.predict(train_data_x)
from sklearn.metrics import log_loss
import numpy as np
y_true = np.array(train_data_y['is_trade'], dtype = bool) # 转换为一维数组，二类用bool可表示
y_pred = np.concatenate(((1-train_predict.reshape(-1,1)), train_predict.reshape(-1,1)), axis = 1) # predict 转换为[p0,p1]格式
ll = log_loss(y_true, y_pred)
print('train_predict log_loss:%f'%ll)
```

## 线性回归模型可行性分析

```python
# 函数化训练测试过程
def testlogloss(train_data, test_data, feature_names, label_name):
    train_data_x = train_data[feature_names]
    train_data_y = train_data[[label_name]]
    test_data_x = test_data[feature_names]
    test_data_y = test_data[[label_name]]
    linreg = LinearRegression()
    linreg.fit(train_data_x, train_data_y)
    result = linreg.predict(test_data_x)
    y_true = np.array(test_data_y['is_trade'], dtype = bool) # 转换为一维数组，二类用bool可表示
    y_pred = np.concatenate(((1-result.reshape(-1,1)), result.reshape(-1,1)), axis = 1) # predict 转换为[p0,p1]格式
    ll = log_loss(y_true, y_pred)
    print('test log_loss:%f'%ll)
    return ll
```

```python
def minmaxlogloss(train_data, test_data, feature_names, label_name, nrandomsamples):
    loglosses= [testlogloss(train_data.sample(n = nrandomsamples), test_data, select_feature, 'is_trade') for i in range(20)]
    print('min:%f, max:%f, diff:%f'%(min(loglosses),max(loglosses),max(loglosses)-min(loglosses)))
        
```

### 随机采样一倍特征数分析

```python
print('self test')
train_sample = train_data.sample(n = len(select_feature))
testlogloss(train_sample, train_sample, select_feature, 'is_trade')
print('others test')
minmaxlogloss(train_data, train_data, select_feature, 'is_trade',len(select_feature)*1 )
```
